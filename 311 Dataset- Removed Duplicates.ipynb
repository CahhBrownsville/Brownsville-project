This analysis is with the 311 dataset. Removing duplicate requests and checking top categories again.

```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
rd311 = pd.read_csv("311_Service_Requests_from_2010_to_Present.csv", parse_dates=["Created Date", "Closed Date"])
pd.set_option('display.max_colwidth', -1)
```

```python
rd311
```

```python
rd311.drop_duplicates(subset=["Complaint Type", "Descriptor"], keep='first', inplace=False)
```

```python
rd311.dtypes
```

-----------------

Checking the Complaint Types

```python
sorted_comp = rd311.sort_values("Complaint Type", ascending = True)
sorted_comp
```

```python
rd311["Complaint Type"]
```

```python
rd311["Complaint Type"].value_counts()
```

```python
comtype_values = rd311["Complaint Type"].value_counts().head(10)
comtype_values.plot(kind = "bar")
```

The top 5 complaint types with the most occurences were "HEAT/HOT WATER", "PLUMBING", "UNSANITARY CONDITION", "HEATING", and "GENERAL CONSTRUCTION". Some of these are vague and need clarification. Made a bar chart of the top 10 issues found. 

--------------

Checking the descriptors.

```python
rd311["Descriptor"]
```

```python
rd311["Descriptor"].value_counts()
```

```python
desc_values = rd311["Descriptor"].value_counts().head(10)
desc_values.plot(kind = "bar")
```

The top descriptors are "ENTIRE BUILDING", "APARTMENT ONLY", "HEAT", "PESTS" and "CEILING". Some of these are unclear, need to look into those. Made a bar chart to show the top 10 issues.
